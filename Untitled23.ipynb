```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/heart_attack_prediction_dataset.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Sex'] = df['Sex'].map({'Male': 1, 'Female': 0})

df[['Systolic BP', 'Diastolic BP']] = df['Blood Pressure'].str.split('/', expand=True).astype(float)
```

```python
df = df.drop(columns=['Blood Pressure'])

df['High Blood Pressure'] = ((df['Systolic BP'] >= 140) | (df['Diastolic BP'] >= 90)).astype(int)
```

```python
df_int = df.select_dtypes(include=['int64', 'float64'])
```

```python
correlation_matrix = df_int.corr()
```

```python
correlation_matrix
```

```python
plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Matrix")
plt.show()
```

```python
plt.figure(figsize=(16, 8))
plt.subplot(1, 3, 1)
sns.histplot(df['Age'], kde=True, bins=30, color='teal')
plt.title('Age Distribution')

plt.subplot(1, 3, 2)
sns.histplot(df['BMI'], kde=True, bins=30, color='purple')
plt.title('BMI Distribution')

plt.subplot(1, 3, 3)
sns.histplot(df['Cholesterol'], kde=True, bins=30, color='orange')
plt.title('Cholesterol Distribution')
plt.tight_layout()
plt.show()
```

```python

# 2. Box Plot: Heart Rate and Triglycerides by Heart Attack Risk
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.boxplot(x='Heart Attack Risk', y='Heart Rate', data=df, palette="Set2")
plt.title('Heart Rate by Heart Attack Risk')

plt.subplot(1, 2, 2)
sns.boxplot(x='Heart Attack Risk', y='Triglycerides', data=df, palette="Set2")
plt.title('Triglycerides by Heart Attack Risk')
plt.tight_layout()
plt.show()

```

```python
# 3. Bar Chart: Smoking, Family History, and Sex Counts against Heart Attack Risk
plt.figure(figsize=(18, 6))
plt.subplot(1, 3, 1)
sns.countplot(x='Smoking', hue='Heart Attack Risk', data=df, palette="magma")
plt.title('Smoking vs Heart Attack Risk')

plt.subplot(1, 3, 2)
sns.countplot(x='Family History', hue='Heart Attack Risk', data=df, palette="magma")
plt.title('Family History vs Heart Attack Risk')

plt.subplot(1, 3, 3)
sns.countplot(x='Sex', hue='Heart Attack Risk', data=df, palette="magma")
plt.title('Sex vs Heart Attack Risk')
plt.tight_layout()
plt.show()
```

```python

# 4. Pair Plot: Pairs of continuous variables for clustering (Age, BMI, Cholesterol, Heart Rate)
sns.pairplot(df, vars=['Age', 'BMI', 'Cholesterol', 'Heart Rate'], hue='Heart Attack Risk', palette="coolwarm")
plt.suptitle('Pair Plot of Continuous Variables', y=1.02)
plt.show()
```

```python

# 5. Heatmap of Country vs Continent count (Optional visualization for region-based analysis)
plt.figure(figsize=(12, 8))
region_data = df.groupby(['Continent', 'Country']).size().unstack(fill_value=0)
sns.heatmap(region_data, cmap='YlGnBu', annot=False, cbar=True)
plt.title('Heatmap of Country by Continent')
plt.show()
```
